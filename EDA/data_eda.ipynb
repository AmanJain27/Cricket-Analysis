```python
corr = df[['boundaries per innings', 'SR', 'Strike Rate without boundaries', 'HS Numeric', 'Ave', 'Avg runs per year', 'Total Years', 'Arithmatic Average of Total Runs by Inns Played', 'Boundary Ave per Inns', '100\'s per innings']].corr()
corr
```

```python
import seaborn as sns
cmap=sns.diverging_palette(220, 15, 80, as_cmap=True)
hm = sns.heatmap(corr, cmap=cmap, linewidths=.5)
hm.set(title='Correlation Matrix of Data')
```

```python
chart = sns.barplot(x=df['Country'].value_counts().index, y=df['Country'].value_counts())
chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
plt.show()
```

```python
chart1 = sns.barplot(x=df['Player'][:30], y=df['NO'][:30])
chart1.set_ylabel("Not Outs")
chart1.set_xlabel("Players")
chart1.set_xticklabels(chart1.get_xticklabels(), rotation=90)
plt.show()
chart2 = sns.barplot(x=df['Player'][30:60], y=df['NO'][30:60])
chart2.set_ylabel("Not Outs")
chart2.set_xlabel("Players")
chart2.set_xticklabels(chart2.get_xticklabels(), rotation=90)
plt.show()
chart3 = sns.barplot(x=df['Player'][60:91], y=df['NO'][60:91])
chart3.set_ylabel("Not Outs")
chart3.set_xlabel("Players")
chart3.set_xticklabels(chart3.get_xticklabels(), rotation=90)
chart3.set_title("Number of Not outs")
plt.show()
```

```python
df_yaxes = df[['Inns', 'NO', 'Runs', 'boundaries', 'Ave', 'Arithmatic Average of Total Runs by Inns Played', 'HS Numeric', 'SR', 'Avg runs per year', 'Boundary Ave per Inns', 'Total Years', '100', '50', '0','Avg runs per year']]
```

```python
for y in df_yaxes:
    fig, axes = plt.subplots(1, 3, figsize=(20, 5), sharey=True)
    fig.suptitle(f'Number of {y}')
    df1 = df.sort_values(by=[y], ascending=False)
    x = df1['Player']
    chart1 = sns.barplot(ax=axes[0], x=x[:30], y=df1[y][:30])
    chart1.set_ylabel(y)
    chart1.set_xlabel("Players")
    chart1.set_xticklabels(chart1.get_xticklabels(), rotation=90)
    axes[0].set_title('first slot of players')
    
    chart2 = sns.barplot(ax=axes[1], x=x[30:60], y=df1[y][30:60])
    chart2.set_ylabel(y)
    chart2.set_xlabel("Players")
    chart2.set_xticklabels(chart2.get_xticklabels(), rotation=90)    
    axes[1].set_title('second slot of players')
    
    chart3 = sns.barplot(ax=axes[2], x=x[60:91], y=df1[y][60:91])
    chart3.set_ylabel(y)
    chart3.set_xlabel("Players")
    chart3.set_xticklabels(chart3.get_xticklabels(), rotation=90)    
    axes[2].set_title('third slot of players')
    
    plt.show()
```
